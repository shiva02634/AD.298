```python
taxi=sns.load_dataset('taxis')
taxi
```

```python
sns.barplot(x='payment',y='total',data=taxi,color='red',estimator='std')
plt.show()

plt.tight_layout()
```

```python
sns.barplot(x='payment',y='total',data=taxi,color='blue',estimator='mean')
plt.show()

plt.tight_layout()
```

```python
sns.histplot(x=taxi['fare'],color='blue')
plt.show()
```

```python
sns.histplot(x=taxi['fare'],color=['blue','red'],hue=taxi['payment'])
plt.show()
```

```python
sns.histplot(x=taxi['fare'],palette=['g','y'],hue=taxi['payment'])
plt.show()
```

```python
sns.distplot(x=taxi['fare'],hist=True,kde=False)
plt.show()
```

```python
sns.distplot(x=taxi['fare'],hist=False,kde=True)
plt.show()
```

```python
sns.distplot(x=taxi['fare'],hist=True,kde=True)
plt.show()
```

```python
sns.boxplot(x=taxi['fare'],color='purple')
plt.show()
```
